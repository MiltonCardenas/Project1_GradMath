## HOMEWORK  

### 1. Goal of the Analysis
The goal of this study is to answer the following questions:

#### Single-Factor Questions:
-   Did people in Pennsylvania who spent more time at work have a higher likelihood of experiencing fatal outcomes during 2020?
-   Were Pennsylvania workers in the food and agriculture sector prone to experiencing fatal outcomes in 2020, as proposed in a previous study conducted by Chen et al (2021) for the state of California?

#### Combined-Factor Question:
-   Were Pennsylvania workers in the food and agriculture sector who spent more time at work more likely to experience fatal outcomes in 2020?

### 2. Summary of the Study Variables:
Measurement: Deaths per county (in Pennsylvania).
Factor 1: Mean Work Time per county (in Pennsylvania).
Factor 2: Workers by occupational sector per county (with an emphasis on the food and agriculture sector in Pennsylvania).


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Extract Pennsylvania data of Work Time
df = pd.read_csv('2020_MeanWorkTime.csv') 
df['state'] = df['NAME'].str.split(', ').str[-1]
df['county'] = df['NAME'].str.split(', ').str[0]
pennsylvania_Timedata = df[df['state'] == 'Pennsylvania']
counties2 = pennsylvania_Timedata['county']
total_hours = pennsylvania_Timedata['B23020_001E']

# Extract Pennsylvania data of COVID
df2 = pd.read_csv('us-counties-2020.csv') 
pennsylvania_COVIDdata = df2[df2['state'] == 'Pennsylvania']
counties = pennsylvania_COVIDdata['county']
deaths = pennsylvania_COVIDdata['deaths']

# Extract Pennsylvania data of Indutry type
df3 = pd.read_csv('Industry_Dataset2020.csv')
df3['state'] = df3['NAME'].str.split(', ').str[-1]
df3['county'] = df3['NAME'].str.split(', ').str[0]
pennsylvania_Workdata = df3[df3['state'] == 'Pennsylvania']
pennsylvania_Workdata['Total_Workers'] = df3[['S2404_C01_002E', 'S2404_C01_003E', 'S2404_C01_004E']].sum(axis=1)
counties3 = pennsylvania_Workdata['county']
Workers = pennsylvania_Workdata['Total_Workers']

# Plot configuration

fig, ax = plt.subplots(3, 1, figsize=(20, 12))
ax[0].bar(counties, deaths, color='hotpink', label="Data per county")
ax[0].set_title('Pennsylvania - Deaths in 2020 due to COVID Panemic')
ax[0].set_xlabel('County')
ax[0].set_ylabel('Deaths')
ax[0].set_yticks([0, 500, 1000, 2000, 2500])
ax[0].tick_params(axis='x', rotation=90)

ax[1].bar(counties2, total_hours, color='grey', label="Data per county")
ax[1].set_title('Pennsylvania - Average hours worked per week in 2020')
ax[1].set_xlabel('County')
ax[1].set_ylabel('Mean Work Time') 
ax[1].set_yticks([20, 30, 40])
ax[1].tick_params(axis='x', rotation=90)

ax[2].bar(counties3, Workers, color='gold', label="Data per county")
ax[2].set_title('Pennsylvania - Workers in Agricultural and Food sector in 2020')
ax[2].set_xlabel('County')
ax[2].set_ylabel('Number of Workers')
ax[2].tick_params(axis='x', rotation=90)

plt.autoscale(enable=True, axis='y')
plt.tight_layout()                                                                  # Ensures labels fit nicely in the plot
plt.show()
```
